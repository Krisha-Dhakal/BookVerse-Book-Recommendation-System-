# Model building

## Popularity based recommendation system

```python
ratings.merge(books, on='ISBN').shape
```

```python
ratings_with_name=ratings.merge(books, on='ISBN')
```

```python
ratings_with_name
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
print(ratings_with_name['Book-Rating'].dtype)
```

```python
non_numeric_ratings = ratings_with_name[pd.to_numeric(ratings_with_name['Book-Rating'], errors='coerce').isna()]
print(non_numeric_ratings)
```

```python
# Ensure all values in 'Book-Rating' are numeric, converting non-numeric values to NaN
ratings_with_name['Book-Rating'] = pd.to_numeric(ratings_with_name['Book-Rating'], errors='coerce')

# Drop rows with NaN values in 'Book-Rating' after conversion
ratings_with_name.dropna(subset=['Book-Rating'], inplace=True)

# Check the data type again
print(ratings_with_name['Book-Rating'].dtype)
```

```python
# Compute the mean rating for each book
avg_rating_df = ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()

# Rename the 'Book-Rating' column to 'avg_rating'
avg_rating_df.rename(columns={'Book-Rating': 'avg_rating'}, inplace=True)

avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
```

```python
popular_df
```

# collaborative fillerting based recommendation system

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
educated_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(educated_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    try:
        # Find the index of the book in the pivot table
        index = np.where(pt.index == book_name)[0][0]

        # Sort books by similarity score
        similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x: x[1], reverse=True)[1:6]

        # Collect recommended books
        recommendations = []
        for i in similar_items:
            recommended_book = pt.index[i[0]]
            # Fetch book details
            book_details = books[books['Book-Title'] == recommended_book].iloc[0]
            recommendations.append({
                'Title': book_details['Book-Title'],
                'Author': book_details['Book-Author'],
                'Image-URL': book_details['Image-URL-M']
            })

        return recommendations

    except IndexError:
        print(f"The book '{book_name}' is not found in the dataset.")
        similar_books = find_similar_books(book_name)

        if similar_books:
            print("*********************************************")
            print(f"Did you mean one of these books?")
            for i, similar_book in enumerate(similar_books, start=1):
                print(f"{i}: {similar_book}")
        else:
            print(f"No similar books found. Please try a different book name.")
        return []

def find_similar_books(book_name):
    similar_books = [b for b in pt.index if book_name.lower() in b.lower()]
    return similar_books[:10] if similar_books else []
```
